```python
import pandas as pd
import os
```

```python
os.listdir()
```

```python
filename = 'abandoned_animal_data_with_pop.csv'
```

```python
df = pd.read_csv(filename)
df.head()
```

```python
df.sort_values(by = ['기간', 'animal', '자치구', '계'])
```

```python
df[(df['기간'] == 2009) & (df['자치구'] == '강남구')]
```

```python
result = pd.melt(df, id_vars=['기간', 'animal', '자치구', '계', '인구총계', '수의사합계'], value_vars=['인도(주인)', '입양분양', '폐사안락사', '계류기증'], var_name=['분류'])
result
```

```python
result[(result['기간'] == 2010) & (result['자치구'] == '은평구')]['value'].sum()
```

```python
result[(result['기간'] == 2009) & (result['자치구'] == '강남구')].sort_values(['animal'])
```

```python
sum_by_categories = result.groupby(['기간', '자치구', '분류'], as_index = False)['value'].sum()
sum_by_categories = sum_by_categories.rename(columns = {'value' : '분류별 합계'})
sum_by_categories
```

```python
result = pd.merge(result, sum_by_categories, on = ['기간', '자치구', '분류'])
result
```

```python
result.to_csv('abandoned_animal_data_with_pop2.csv',index=False)
```
